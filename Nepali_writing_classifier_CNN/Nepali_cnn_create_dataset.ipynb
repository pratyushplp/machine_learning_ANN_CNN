```python
# load own handwriting
dataset = []
##img_size=20
count=0

# load the png image data
def create_training_data():
    os.chdir('D:/')
    for image_file_name in glob.glob('ML/nepali_characters/*/*/*.jpg'):
        
        label = int(image_file_name[-14:-11])    
        try:
            img_array=cv2.imread(image_file_name,cv2.IMREAD_GRAYSCALE)
            ##new_array=cv2.resize(img_array,(img_size,img_size))
        except Exception as e:
            print("warning image is broken")
        ##img_temp=new_array.reshape(400)
        #normalization
        img_data = (img_array/ 255)
        dataset.append([img_data,label])
        #record = np.append(img_data,label)
        #dataset.append(record)
    return dataset
data=create_training_data()
print(data[1])
```

```python
data=create_training_data()
data=np.array(data)
print(data.shape)
#print(data[1])
```
